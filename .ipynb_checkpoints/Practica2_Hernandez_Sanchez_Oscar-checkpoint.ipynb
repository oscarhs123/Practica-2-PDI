```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
# comando de Jupyter para que las imagenes se muestren automaticamente
%matplotlib inline

'Aquí inicia los Ejercicios 1 y 2'
#Ejercicio 1: Leer imagen de 1024 x 1024
plt.rcParams['image.cmap'] = 'gray'
rosa_original = io.imread("imagenes/rosa1024.jpg") #Lee la imagen
rosa_original = rosa_original/255.0 #nomalizamos la imagen
print(rosa_original.shape) #imprimimos datos de la imagen
'Esto es parte del Ej 2'
plt.imshow(rosa_original) #mostramos la imagen
plt.title("Imagen original")
resolucion = rosa_original.shape
```

```python
#Ejercicio de reducción a 512
rosa_512 = np.ones(((int(resolucion[0]/2.0)),(int(resolucion[1]/2.0)))) #creamos una nueva imagen del tamaño deseado
print(rosa_512.shape) 
plt.imshow(rosa_512)
rosa_512_b = rosa_512.shape

'Esto es parte del Ej 2'
for y in range(0, rosa_512_b[0]):
    for x in range(0,rosa_512_b[1]):
        rosa_512[y][x] = rosa_original[y*2][x*2]
plt.imshow(rosa_512) 
plt.title("Imagen original reducida a la resolución 512")
```

```python
#Ejercicio de reducción a 256
rosa_256 = np.ones(((int(resolucion[0]/4.0)),(int(resolucion[1]/4.0)))) #creamos una nueva imagen del tamaño deseado
print(rosa_256.shape) 
plt.imshow(rosa_256)
rosa_256_b = rosa_256.shape

'Esto es parte del Ej 2'
for y in range(0, rosa_256_b[0]):
    for x in range(0,rosa_256_b[1]):
        rosa_256[y][x] = rosa_original[y*4][x*4]
plt.imshow(rosa_256)
plt.title("Imagen original reducida a la resolución 256")
```

```python
'Aquí inicia ejercicio 3'

# Función para aplicar zoom a imagenes recibe la imagen a ampliar y la resolución a la que se amplía 
def miZoom(img, res):
    imagen_ampliada = np.zeros(res)  # Creamos una nueva imagen de tamaño deseado
    
    # Hacemos la proporción a la nueva resolución
    y_zoom = img.shape[0] / res[0]
    x_zoom = img.shape[1] / res[1]
    
    # Aplicamos vecino más cercano
    for y in range(res[0]): # Recorremos ambas dimensiones de la imagen
        for x in range(res[1]):
            vecino_y = min(int(y * y_zoom), img.shape[0] - 1) #Calculamos Vecino más cercano a y
            vecino_x = min(int(x * x_zoom), img.shape[1] - 1) #Calculamos Vecino más cercano a x
            imagen_ampliada[y, x] = img[vecino_y, vecino_x] 
    
    return imagen_ampliada
```

```python
# Aplicamos el zoom a 1024x1024 usando nuestra funcion miZoom
rosa_512_ampliada = miZoom(rosa_512, (1024, 1024))

# Mostramos la imagen de 512 ampliada a 1024x1024
plt.imshow(rosa_512_ampliada)
plt.title("Imagen de resolución 512 ampliada a 1024")
```

```python
# Aplicamos el zoom a 1024x1024 usando nuestra funcion miZoom
rosa_256_ampliada = miZoom(rosa_256, (1024, 1024))
# Mostramos la imagen de 256 ampliada a 1024x1024
plt.imshow(rosa_256_ampliada)
plt.title("Imagen de resolución 256 ampliada a 1024")
```

```python
'Ejercicio 4'
#Funcion que Cuantiza una imagen dada a un número específico de niveles de intensidad.
#Parámetros:
#img: La imagen original que se va a cuantizar.
#cuantizar: El número de niveles de intensidad a los que se va a cuantizar la imagen.
#Return:
#La imagen cuantizada con el número especificado de niveles de intensidad.

def cuantizador(img, niveles):
    
    #Multiplicamos por el numero de niveles a cuantizar
    #con floor redondeamos para obtener valores entre 0 y <niveles>
    #dividimos para escalar entre 0 y <niveles>
    imagen_cuantizada = np.floor(img * niveles) / niveles 
    return imagen_cuantizada
```

```python
'Aquí inicia el ejercicio 6'
import random
#Aquí solo leemos nuestra imagen de 64 x 64
plt.rcParams['image.cmap'] = 'gray'
imagen_64 = io.imread("imagenes/camila_64.jpg") #Lee la imagen
imagen_64 = imagen_64 / 255.0 #nomalizamos la imagen
print(imagen_64.shape) #imprimimos datos de la imagen
plt.imshow(imagen_64) #mostramos la imagen
plt.title("Imagen de 64x64 px")
```

```python
# Función para encontrar píxeles adyacentes
def adyacentes(pixeles):
    adyacentes4 = [] #Arreglo para guaradar los vecinos de la 4-adyacencia
    adyacentes8 = [] #Arreglo para guaradar los vecinos de la 8-adyacencia
    for x, y in pixeles:
        # Píxeles 4-adyacentes
        adyacentes4.append([(x, y - 1), (x - 1, y), (x + 1, y), (x, y + 1)]) #Agregamos los vecinos 4-adyacentes de cada pixel

        # Píxeles 8-adyacentes
        adyacentes8.append([(x - 1, y - 1), (x, y - 1), (x + 1, y - 1),(x - 1, y),
                             (x + 1, y),(x - 1, y + 1), (x, y + 1), (x + 1, y + 1)]) #Agregamos los vecinos 8-adyacentes de cada pixel

    return adyacentes4, adyacentes8 #devolvemos las 4 y 8 adyacencias
```

```python
coordenadas_aleatorias = [(random.randint(0, 63), random.randint(0, 63)) for _ in range(5)]
# Encontrar píxeles adyacentes para las coordenadas aleatorias
mis_4_adyacentes, mis_8_adyacentes = adyacentes(coordenadas_aleatorias) #obetnesmos los 4 y 8 adyacentes de las coordenadas aleatorias

#Imprimimos las coordenadas aleatorias y sus 4 y 8 adyacencias
for i in range(5):
    print(f"Coordenada Aleatoria {i + 1}: {coordenadas_aleatorias[i]}")
    print(f"Píxeles 4-adyacentes: {mis_4_adyacentes[i]}")
    print(f"Píxeles 8-adyacentes: {mis_8_adyacentes[i]}")
    print("\n")



'Aquí inicia el ejercicio 7'
#Marcamos la coordenada centro de Blanco
imagen_64[coordenadas_aleatorias[0]] = 1
imagen_64[coordenadas_aleatorias[1]] = 1
imagen_64[coordenadas_aleatorias[2]] = 1
imagen_64[coordenadas_aleatorias[3]] = 1
imagen_64[coordenadas_aleatorias[4]] = 1
    

#Marcar en la imagen las 4-adyacencias
for x, y in coordenadas_aleatorias:
    # Píxeles 4-adyacentes
    adyacentes_4 = [(x, y - 1), (x - 1, y), (x + 1, y), (x, y + 1)]
    
    # Marcar los vecinos 4-adyacentes con gris (0.5)
    for x_1, y_1 in adyacentes_4:
        if 0 <= x_1 < 64 and 0 <= y_1 < 64: #Establecemos limite para evitar que se salga del rango
            imagen_64[x_1, y_1] = 0.5
            
#Marcar en la imagen las 8-adyacencias
for x, y in coordenadas_aleatorias:
    # Píxeles 8-adyacentes (Solo esquinas ya que los 4-adyacentes ya los marcamos previamente)
    adyacentes_8 = [(x - 1, y - 1), (x - 1, y + 1), (x + 1, y - 1), (x + 1, y + 1)]
    
    # Marcar los vecinos 8-adyacentes con negro
    for x_1, y_1 in adyacentes_8:
        if 0 <= x_1 < 64 and 0 <= y_1 < 64: #Establecemos limite para evitar que se salga del rango
            imagen_64[x_1, y_1] = 0

plt.imshow(imagen_64)
plt.title("Imagen de 64x64 con 5 pixeles aleatorios marcados con Blanco con: \n" + "Sus 4 y 8 adyacencias (con gris y negro)")
```

```python
'Aquí inicia el ejercicio 8'

# Tomamos un pixel de nuestras coordenadas
pixel_elegido = coordenadas_aleatorias[1]

# Calculamos la distancia City Block y Chessboard para la distancia a los demas pixeles
for i in range(5):
    pixel_actual = coordenadas_aleatorias[i]
    city_block = abs(pixel_elegido[0] - pixel_actual[0]) + abs(pixel_elegido[1] - pixel_actual[1]) #abs calcula el valor absoluto
    chessboard = max(abs(pixel_elegido[0] - pixel_actual[0]), abs(pixel_elegido[1] - pixel_actual[1])) #max devuelve el valor máximo
    
    print(f"Distancia City Block al pixel {i + 1}: {city_block}")
    print(f"Distancia Chessboard al pixel {i + 1}: {chessboard}")
    print("\n")
```
